```python
def import_data(filename):
    DataSet = list()
    import_data = open(filename, 'r').read().split('\n')
    for i in range(len(import_data)):
        data_row = import_data[i].split(',')
        DataSet.append(data_row[0:])
    return DataSet
```

```python
dataset = import_data(data_path)
print(dataset)
vector = import_data(vector_path)
print(vector)
```

```python
def str_column_to_float(dataset, column):
    for row in dataset:
        row[column] = float(row[column])
```

```python
def dataset_to_float(filename):
    dataset = import_data(filename)
    FloatedData = dataset
    for i in range(1,len(FloatedData[0])):
        str_column_to_float(FloatedData, i)
    return FloatedData
```

```python
dataset_to_float(vector_path)
dataset_to_float(data_path)
```

```python
def dataset_to_panda(filename):
    NewDataSet = dataset_to_float(filename)
    Lables = [lable[0] for lable in NewDataSet]
    Columns = ['Name']+[i for i in range(1,len(NewDataSet[0]))]
    DataFrame = pd.DataFrame(data=NewDataSet, columns=Columns)
    return DataFrame
```

```python
import pandas
def fisher_poj(probka1_start,probka1_stop,probka2_start,probka2_stop):
    
    print('Probki')
    probka1 = Panda_data[probka1_start:probka1_stop]
    probka2 = Panda_data[probka2_start:probka2_stop]
    
    display(probka1)
    display(probka2)
    
    liczba_cech1 = len(probka1.columns)-1
    liczba_cech2 = len(probka1.columns)-1
    
    avg_probka1 = probka1.mean()
    avg_probka2 = probka2.mean()
    

    avg1= pd.DataFrame(avg_probka1).transpose()
    avg2= pd.DataFrame(avg_probka2).transpose()
    
    print('Wartosci srednie')
    
    display(avg1,avg2)
    
    print('Odchylenia')
    
    odchylenie1 = probka1.std(axis=1)   # Mozna Var i pierw, ta działa identycznie
    odchylenie2 = probka2.std(axis=1)
    
    odchylenie1 = pd.DataFrame(odchylenie1)
    odchylenie2 = pd.DataFrame(odchylenie2)
    
    display(odchylenie1,odchylenie2)
    
    print('Odchylenia odejmowanko i moduł')

    odchylenie = odchylenie1.reset_index(drop=True) - odchylenie2.reset_index(drop=True)
    odchylenie = odchylenie.abs()
    
    display(odchylenie)
    
    print('Odejmowanko i moduł')
    odejmowanko = avg1.subtract(avg2, axis=1)
    odejmowanko = odejmowanko.abs()
    odejmowanko = pd.DataFrame(odejmowanko)
    
    display(odejmowanko)

    
    print('Edycja odejmowanek do dzielenia')

    odejmowanko = pd.concat([odejmowanko]*(probka1_stop-probka1_start),axis = 0, ignore_index=True)
    a = probka1.index.values
    a = a.tolist()
    a = { i : a[i] for i in range(0, len(a) ) }
    print(a)
    odejmowanko.rename(mapper = a, inplace = True)

    
    display(odejmowanko)

    print('Edycja odchylenia')
    
    odchylenie = pd.concat([odchylenie]*len(probka1.columns),axis = 1,ignore_index = True)
    odchylenie = odchylenie.drop(labels =  0, axis = 1)

    
    display(odchylenie)
    
    
    print('Dzielenie')
    odejmowanko = odejmowanko.reset_index(drop=True)
    odchylenie = odchylenie.reset_index(drop=True)
    
    dzielonko = odejmowanko.div(odchylenie)
    display(dzielonko)
    print('Maxy - Max->Max_index')
    
    Maks = dzielonko.max(axis=1)
    Maks = pd.DataFrame(Maks)
    
    Maks_index = dzielonko.idxmax(axis=1)
    Maks_index = pd.DataFrame(Maks_index)

    display(Maks,Maks_index)

    return Maks,Maks_index

```

<h1>Fisher multi<h1>

```python
def fisher_mul(probka1_start,probka1_stop,probka2_start,probka2_stop,cechy):
    
    
    print('Probki')
    probka1 = Panda_data[probka1_start:probka1_stop]
    probka2 = Panda_data[probka2_start:probka2_stop]
    
    probka1 = probka1.loc[probka1_start:probka1_stop,cechy]
    probka2 = probka2.loc[probka2_start:probka2_stop,cechy]
    
    display(probka1)
    display(probka2)
    
    liczba_cech1 = len(probka1.columns)-1
    liczba_cech2 = len(probka1.columns)-1
    
    avg_probka1 = probka1.mean()
    avg_probka2 = probka2.mean()
    

    avg1= pd.DataFrame(avg_probka1).transpose()
    avg2= pd.DataFrame(avg_probka2).transpose()
    
    print('Wartosci srednie')
    
    display(avg1,avg2)
    
    print('Odejmowanko i moduł')
    odejmowanko = avg1.subtract(avg2, axis=1)
    odejmowanko = odejmowanko.abs()
    odejmowanko = pd.DataFrame(odejmowanko)
    
    display(odejmowanko)

    
    print('Edycja odejmowanka do dzielenia')

    odejmowanko = pd.concat([odejmowanko]*(probka1_stop-probka1_start),axis = 0, ignore_index=True)
    a = probka1.index.values
    a = a.tolist()
    a = { i : a[i] for i in range(0, len(a) ) }
    odejmowanko.rename(mapper = a, inplace = True)
    
    
    display(odejmowanko)
    
    print('Kowariancje')

    Kow1 = probka1.cov()
    Kow2 = probka2.cov()
    
    display(Kow1,Kow2)
    
    print('Suma Kowariancji')
    Kow1 = Kow1.to_numpy()
    Kow2 = Kow2.to_numpy()
    
    
    Kow1 = np.linalg.det(Kow1)
    Kow2 = np.linalg.det(Kow2)
    
    print(Kow2,Kow1)
          
    suma = (Kow1+Kow2)
    print('Suma =',suma)
    
    print('')
    print('Fisher')
    Fisherro = odejmowanko.div(suma)
    display(Fisherro)
    
    print('Maxy - Max->Max_index')
    
    Maks = Fisherro.max(axis=1)
    Maks = pd.DataFrame(Maks)
    
    Maks_index = Fisherro.idxmax(axis=1)
    Maks_index = pd.DataFrame(Maks_index)

    display(Maks,Maks_index)

    
    print('')
    print('Return')
    
    return (Maks_index)
```
