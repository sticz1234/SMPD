<h1>NN<h1>

```python
def import_data(filename):
    DataSet = list()
    import_data = open(filename, 'r').read().split('\n')
    for i in range(len(import_data)):
        data_row = import_data[i].split(',')
        DataSet.append(data_row[0:])
    return DataSet
```

```python
dataset = import_data(data_path)
print(dataset)
vector = import_data(vector_path)
print(vector)
```

```python
def str_column_to_float(dataset, column):
    for row in dataset:
        row[column] = float(row[column])
```

```python
def dataset_to_float(filename):
    dataset = import_data(filename)
    FloatedData = dataset
    for i in range(1,len(FloatedData[0])):
        str_column_to_float(FloatedData, i)
    return FloatedData
```

```python
dataset_to_float(vector_path)
dataset_to_float(data_path)
```

```python
def dataset_to_panda(filename):
    NewDataSet = dataset_to_float(filename)
    Lables = [lable[0] for lable in NewDataSet]
    Columns = ['Name']+[i for i in range(1,len(NewDataSet[0]))]
    DataFrame = pd.DataFrame(data=NewDataSet, columns=Columns)
    return DataFrame
```

```python
Panda_data = dataset_to_panda(data_path)
Panda_vector = dataset_to_panda(vector_path)
```

<h1>KNN<h1>

```python
n = 3  #LICZBA SASIADOW !!!!!!
Neighbours = DistanceFrame.apply(lambda x: pd.Series(x.nsmallest(n).index))
```

```python
Neighbours
```

```python
minilistknn = np.array(Neighbours).tolist()
```

```python
flat_minilistknn= []
for sublist in minilistknn:
    for item in sublist:
        flat_minilistknn.append(item)
```

```python
lista_predykcja = list(Panda_data["Name"].iloc[flat_minilistknn])
```

```python
split_lista_predykcja = [lista_predykcja[i:i + n] for i in range(0, len(lista_predykcja), n)] #splitujemy liste na  podlisty po n sasiadow
```
